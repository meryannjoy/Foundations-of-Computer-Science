```python
import pandas as pd
import os
```

```python
import os
cartellavideos='/Users/meryannjoy/Desktop/DATA SCIENCE/FOCS/GitHub/Foundations-of-Computer-Science/videos'
# Usa list comprehension per ottenere solo i file che finiscono in .csv
# e crea la lista che stai cercando di usare
livideos = []
for f in os.listdir(cartellavideos): 
    if f.endswith('.csv'):
        livideos.append(f)

print(livideos)
```

```python
df_temp = pd.read_csv('/Users/meryannjoy/Desktop/DATA SCIENCE/FOCS/GitHub/Foundations-of-Computer-Science/videos/CAvideos.csv', sep=',') 
print(df_temp)
```

```python
# Lista vuota per contenere i DataFrame individuali
lista_df = []
import os
# Iterazione su ogni nome di file nella lista 'videos'
for nome_file in livideos:
    
    # 1. Costruisce il percorso completo: 'videos/nome_file.csv'
    # Questo risolve l'errore FileNotFoundError
    percorso_completo = os.path.join('/Users/meryannjoy/Desktop/DATA SCIENCE/FOCS/GitHub/Foundations-of-Computer-Science/videos/CAvideos.csv', nome_file) 
    
    # 2. Estrae lo stato (le prime due lettere del nome del file)
    stato = nome_file[:2]
    
    # 3. Legge il file CSV (devi specificare il percorso completo!)
    # Assumiamo il separatore standard ','
    df_temp = pd.read_csv(percorso_completo, sep=',') 
    
    # 4. Aggiunge la nuova colonna 'stato' al DataFrame temporaneo
    df_temp['stato'] = stato
    
    # Aggiunge il DataFrame alla lista
    lista_df.append(df_temp)

# 5. Concatena tutti i DataFrame in uno unico
df_finale = pd.concat(lista_df, ignore_index=True)

# 6. Output finale
print("-" * 50)
print("DataFrame Finale Creato con Successo!")
print(f"Righe totali: {df_finale.shape[0]}")
print(f"Verifica stati aggiunti: {df_finale['stato'].unique()}")
print("-" * 50)
df_finale.head() # Visualizza le prime righe del risultato
```
